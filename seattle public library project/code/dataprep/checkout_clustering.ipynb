# Clustering

### Problem Definition
Can we identify meaningful clusters of non-fiction library books based on their attributes?

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import StandardScaler

import scipy
import scipy.cluster.hierarchy as sch
import collections

sns.set_style('whitegrid')
```

The data will be a small subset of the non-fiction checkouts in January at the Seattle Public Library in 2016. Bigger sets were taking too long and proving nearly impossible to analyze.

```python
df = pd.read_csv('../../data/processed/df_cluster.csv')
print(df.columns)
df.head()
```

```python
col = 'deweySubject'
df_dummies = pd.get_dummies(df[col])
df_dummies.columns = [str(col)+'_'+str(c) for c in df_dummies.columns]
#df_train = pd.concat([df, df_dummies], axis=1)
```

```python
df_dummies = pd.get_dummies(df['deweySubject'], prefix='subject')
```

```python
df_dummies
```

```python
df = pd.concat([df, df_dummies], axis=1)
```

```python
df = df.sample(1000)
```

```python
df_sam = df.copy()
```

### Use numerical columns and subjects

```python
# select the columns
X_columns = ['Age', 'Count', 'Child', 'Teen', 'subject_Arts & Recreation',
       'subject_Computer Science', 'subject_History & Geography',
       'subject_Language', 'subject_Literature',
       'subject_Philosophy & Psychology', 'subject_Pure Science',
       'subject_Religion', 'subject_Social Sciences', 'subject_Technology']

# normalize the data
for col in X_columns:
    df[col] = StandardScaler().fit_transform(df[col].values.reshape(-1, 1))
```

```python
k = 9
color_threshold = 45
d = sch.distance.pdist(df[X_columns])
Z= sch.linkage(d, method = 'ward') # minimize within cluster variation
T = sch.fcluster(Z, k, 'maxclust')
P = sch.dendrogram(Z, color_threshold=color_threshold)
plt.show()

print(set(T))
print(collections.Counter(T))

df_results = df_sam.copy()
df_results['cluster'] = T
df_sam['cluster'] = T
```

```python
# Analyze the results
for cluster in set(T):
    print((cluster), (len(df_results[df_results['cluster']==cluster]['Title'])))
    print(sorted(list(df_results[df_results['cluster']==cluster]['Title'])))
```

```python
# Analyze the centroids
df_results.groupby('cluster').mean().round(2).T
```

# Results
The results don't seem very interesting with the subjects added. The clusters are mostly influenced by the subjects of the books so there is basically a cluster for each subject. The average age of the clusters is slightly interesting since there aren't any clusters that are extremely old or really new. The same can be said about the average count as it is pretty consistent across the clusters. Next we'll try removing the subject dummies and only use the numerical columns to see if anything else can be discoved.

```python
# select the columns
X_columns = ['Age', 'Count', 'Child', 'Teen']
```

```python
k = 12
color_threshold = 5
d = sch.distance.pdist(df[X_columns])
Z= sch.linkage(d, method = 'ward') # minimize within cluster variation
T = sch.fcluster(Z, k, 'maxclust')
P = sch.dendrogram(Z, color_threshold=color_threshold)
plt.show()

print(set(T))
print(collections.Counter(T))

df_results = df_sam.copy()
df_results['cluster'] = T
df_sam['cluster'] = T
```

```python
# Analyze the results
for cluster in set(T):
    print((cluster), (len(df_results[df_results['cluster']==cluster]['Title'])))
    print(sorted(list(df_results[df_results['cluster']==cluster]['Title'])))
```

```python
# Analyze the centroids
df_results.groupby('cluster').mean().round(2).T
```

```python
# Analyze the results
for col in X_columns:
    print(col)
    for cluster in set(T):
        plt.hist(df_results[df_results['cluster']==cluster][col], label=str(cluster), alpha=0.3, bins=20)
    #plt.legend()
    plt.show()
```

## Results
Removing the subjects gives some more noteable clusters based on the numerical values. There are books that are significantly older and are checked out infrequently and books that are new and very popular. They seem to be mostly grouped by age. There are also two clusters that are strictly based on being for teens or children. Cluster 5 and 7 are both very new and popular but only contain 3 books between them. It appears that while they are both very new they were made separate categories because of a 40+ count difference.

In the end, the clustering would take a lot more tuning and ideally more numerical columns like number of page numbers, time borrowed, average requests, etc. The few categories available weren't particularly helpful for grouping. Another very key factor is the number of titles being grouped. The sample of 1000 is not enough and at the same time it is very hard to look at such big clusters. Additionally, perhaps it would be more useful to compile the clusters by something other than titles.

Clustering seems perfect for library collections but it obviously requires clear goals and deep domain knowledge. This paper from 1995 in the journal "Pattern Regognition" entitled "Knowledge-based clustering scheme for collection management and retrieval of library books" proposes using clustering for "(1) helping the user select an appropriate collection of books in a library which contains the topics of interest; (2) assigning a classification number to a new book; (3) designing a more appropriate and uniform classification scheme for books".
